```python
import geopandas as gpd
import geodatasets

lands_path = geodatasets.get_path("naturalearth.land")
# lands_path = geodatasets.get_path("geoda.nepal")

world = gpd.read_file(lands_path)
kiva_loan_data = gpd.read_file('./data/kiva/kiva_loans.shp')
```

```python
ax = world.plot(figsize=(20,20), color='whitesmoke', linestyle=':', edgecolor='black')
kiva_loan_data.plot(ax=ax, markersize=2, cmap='Greens', scheme='quantiles')
# kiva_loan_data.plot(ax=ax, column='sector', legend=True, cmap='OrRd')
```

```python
ph_loans = kiva_loan_data.loc[kiva_loan_data.country == 'Philippines'].copy()

phl_boundries = gpd.read_file('./data/kiva/Philippines_AL258.kml', driver='KML')
```

```python
ax = phl_boundries.plot(figsize=(12,18), color="whitesmoke", linestyle=':', edgecolor="lightgray")

ph_loans.plot(ax=ax, column='sector', legend=True)
```
