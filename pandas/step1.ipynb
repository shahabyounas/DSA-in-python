```python
reviews['index_backwords'] = range(len(reviews), 0, -1)
reviews['index_backwords']
```

```python
rp_mean = reviews.points.mean()


# reviews.points.map(lambda p: p - rp_mean)


def rememan_points(row):
    row.points = row.points - rp_mean
    return row

reviews.apply(rememan_points, axis='columns')
```

```python
# reviews.price.max(0)
pp_ratio = reviews.price / reviews.points
pp_ratio_max_id = pp_ratio.idxmax()
pp_ratio_max_id
bargain_wine = reviews.loc[pp_ratio_max_id, 'title']
```

```python
    
n_trop = reviews.description.map(lambda d: "tropical" in d).sum()
n_fruity = reviews.description.map(lambda d: "fruity" in d).sum()
descriptor_counts = pd.Series([n_trop, n_fruity], index=['tropical', 'fruity'])
descriptor_counts

reviews.country[reviews.country == 'Canada']
```

```python

# score 95 and higher - 3 Star
# socre 85 and less than 95 - 2 Star
# Any other score 1 star
# Any wine from Canada - 3 Star

def stars(row):
    if row.country == 'Canada':
        return 3
    if row.points >= 95:
        return 3
    if row.points >= 85:
        return 3
    else:
        return 1


star_ratings = reviews.apply(stars, axis="columns")
```

```python

# reviews.groupby('country').points.value_counts()
# reviews.points.value_counts()
# reviews.groupby('points').price.max()

# Explicitly define the grouping cols [['title', 'winery']]
# reviews.groupby('winery')[['title', 'winery']].apply(lambda df: df.title.iloc[0])

# Best wine by country and province
def dfs(df):
    return df.loc[df.points.idxmax()]
# reviews.groupby(['country', 'province'])[['points']].apply(lambda df: df.loc[df.points.idxmax()])
reviews.groupby(['country', 'province'])[['points', 'country']].apply(dfs)
```
