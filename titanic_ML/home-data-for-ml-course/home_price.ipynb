```python

y = train_data['SalePrice']

house_features = ['MSSubClass','MSZoning', 'LotArea', 'MoSold', 'TotRmsAbvGrd', 'Neighborhood', 'OverallQual', 
                  'OverallCond', 'YearBuilt', 'Foundation', 'HouseStyle']

X = train_data[house_features]

X_dumies = pd.get_dummies(X)
```

```python
from sklearn.metrics import mean_absolute_error, root_mean_squared_error
from sklearn.tree import DecisionTreeRegressor

def get_predictions(xInput, y, leaf_nodes =10):
    pricing_model = DecisionTreeRegressor(max_leaf_nodes=leaf_nodes)
    pricing_model.fit(xInput, y)
    predictions = pricing_model.predict(xInput)
    return predictions

def get_mae(xInput, y, max_leaf_nodes = 10):
    preds = get_predictions(xInput, y, max_leaf_nodes)
    return root_mean_squared_error(y, preds)

for max_leaf_nodes in [5, 10, 50, 75, 100, 250, 500, 1000, 5000]:
    mae = get_mae( X_dumies, y, max_leaf_nodes)
    print("Max leaf nodes: %d \t Mean absolute error: %d" %(max_leaf_nodes, mae))


# mean_absolute_error(y, preds)
# root_mean_squared_error(y, preds)
```

```python
from sklearn.ensemble import RandomForestRegressor

text_X = pd.get_dummies(test_data[house_features])
fr_model = RandomForestRegressor(random_state=1)
fr_model.fit(text_X, y)
fr_preds = fr_model.predict(text_X)
mean_absolute_error(y, fr_preds)
```
